```python
import pandas as pd
```

### US COVID cases

```python
cdf = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/09-12-2020.csv')
cdf.head()
```

```python
cdf = cdf.sort_values('Confirmed', ascending=False)
cdf.plot.bar(x='Province_State', y= 'Confirmed', figsize = (24,8))
```
